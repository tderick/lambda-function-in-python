# Import malaria dataset


```python
import pandas as pd
data = pd.read_csv('DatasetAfricaMalaria.csv')
```

# Explore malaria dataset

```python
data.info()
data.head()
data.describe()
```

# List of country that the name start with "C"


```python
filtering_cond = data['Country Name'].apply(lambda country: country.startswith('C'))
data[filtering_cond]
```

# List of country without Malaria case reported

```python
filtering_cond = data['Malaria cases reported'].apply(lambda nbre_malaria_case: nbre_malaria_case==0) 
data[filtering_cond]
```
